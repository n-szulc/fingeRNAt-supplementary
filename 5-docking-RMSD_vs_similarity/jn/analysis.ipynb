```python
dataFileFull = "/home/filips/IIMCB/fingernat/FN1-REVIEWS/R3/5-docking-RMSD_vs_similarity/1-results/collected.tsv"
outdir="/home/filips/IIMCB/fingernat/FN1-REVIEWS/R3/5-docking-RMSD_vs_similarity/1-results/markdown"

outMdFile = outdir + "/analysis.md"


headers = ['PDB', 'metrics', 'Spearman', 'Pearson', 'Kendall', 'RÂ²']
```

```python
def file_append(filename, content):
  f = open( filename, 'a' )
  f.write(content)
  f.close() 

def file_write(filename, content):
  f = open( filename, 'w' )
  f.write(content)
  f.close() 
```

## Loop over groups

```python
def rysujScatter2(x, y, nativeVector, ax):
    g = sns.scatterplot(
        x=x,
        y=y,
        legend=False,
        alpha=0.75,
        ax=ax,
        color="gray",
        edgecolor="black",  # for edge color
        linewidth=0.4,  # line width of spot
        s=15,
    )

    g = sns.scatterplot(
        x=[nativeVector[0]],
        y=[nativeVector[1]],
        legend=False,
        alpha=0.9,
        ax=ax,
        color="red",
        edgecolor="black",  # for edge color
        linewidth=0.4,  # line width of spot
        s=35,
    )

    # linear correlation here + plots
    coef = np.polyfit(x,y,1)
    poly1d_fn = np.poly1d(coef)

    zakres = np.arange(min(x), max(x), 0.1)
    ax.plot(zakres, poly1d_fn(zakres), '--b');
    
        
    plt.subplots_adjust(wspace=0.8, hspace=0.8)

```

```python

#
# THE LOOP OVER PDBS
#


# initiate the output file
file_write(outMdFile, "")

correlationsGlobal = pd.DataFrame(columns=headers)

# loop over the pdbs

grouped = data.groupby(['PDB'])

for pdb, group in grouped:
    print(pdb)
    
    file_append(outMdFile, "\n# %s\n" % (pdb))
    
    group = group.drop('PDB', axis=1)

    rmsd = group['RMSD']

    # statistics
    file_append(outMdFile, "\n\n### Statistics of similarity metrics\n\n")
    file_append(outMdFile, group.describe().to_markdown())

    # parameters - correlation ,r2 etc.
    correlations = pd.DataFrame(columns=headers)
    
    
    # plots initialize
    fig, axes = plt.subplots(2, 5, sharex=False, sharey=False, figsize=(14, 5))
    
    i = -1 # counter
    for metrics in group.drop(['pose', 'RMSD'], axis=1):
        print(metrics)
        i+=1

        # regression
        res = linregress(rmsd, group[metrics])
        r2 = res.rvalue**2

        rmsdAndMetrics = pd.merge(rmsd,
                                  group[metrics],
                                  left_index=True,
                                  right_index=True)
        
        # rank correlations
        corrSpearman = rmsdAndMetrics.corr(method='spearman')[[
            'RMSD'
        ]].iloc[1:].values[0][0]
        corrPearson = rmsdAndMetrics.corr(method='pearson')[[
            'RMSD'
        ]].iloc[1:].values[0][0]
        corrKendall = rmsdAndMetrics.corr(method='kendall')[[
            'RMSD'
        ]].iloc[1:].values[0][0]

        correlations.loc[len(correlations)] = [
            pdb, metrics, corrSpearman, corrPearson, corrKendall, r2
        ]
        
        
        correlationsGlobal.loc[len(correlationsGlobal)] = [
            pdb, metrics, corrSpearman, corrPearson, corrKendall, r2
        ]
        
        #
        # add/draw plots
        #
        
        # coordinates of the native pose
        
#         nativeLine = group[group.pose == 'native']
#         nativeVector = [0, nativeLine[metrics].iloc[0] ]
        
        
#         whichaxY = i % 5
#         whichaxX = int(i / 5)
      
#         rysujScatter2(x=rmsd, y=group[metrics], nativeVector=nativeVector, ax=axes[whichaxX,whichaxY])


    # two last plots - hostograms
    
#     sns.histplot(x=group['tanimoto'], kde=True, ax=axes[1,3], color="gray",)
#     sns.histplot(x=group['tversky'], kde=True, ax=axes[1,4], color="gray",)
#     plt.savefig("%s/imgs/scatterplots-%s.png" % (outdir, pdb), dpi=150, bbox_inches="tight")
    
    file_append(outMdFile, "\n\n### RMSD vs fingerprint similarity\n\n")
    file_append(outMdFile, correlations.to_markdown())
    
    file_append(outMdFile, "\n\n")
    file_append(outMdFile, "![RMSD vs metrics](imgs/scatterplots-%s.png)\n" % (pdb))
    
#     break
```
